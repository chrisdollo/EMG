```python
def extract_features(file_path, sub_num, sampling_rate, window_size, step_size):

    # Load the MATLAB file
    mat_data = scipy.io.loadmat(file_path)
    data = mat_data["processed_data"]
    set_of_all_gesture_by_subject = np.array(data)

    gestures = set_of_all_gesture_by_subject.shape[1]
    repetitions = set_of_all_gesture_by_subject.shape[0]
    sensors = (set_of_all_gesture_by_subject[0][0].shape)[1]

    features_for_gesture_type = []


    # all gestures types
    for i in range(gestures):

        features_for_repetition = []

        # all repetitions of a specific type
        for j in range(repetitions): # shape (28,)

            features_for_signal = []

            # all channels 
            for k in range(sensors):

                # feature Extraction
                
                emg_features, features_names = features_estimation(set_of_all_gesture_by_subject[j][i][:,k],  "Gesture " + str(i+1) + " repetition " + str(j+1) + " channel " + str(k+1)  , sampling_rate, window_size, step_size)
                emg_features = emg_features.to_numpy()
                print(emg_features.shape)                
                
                features_for_signal.append(emg_features[:,0].tolist())
                print(len(features_for_signal))    

            features_for_repetition.append(features_for_signal) 

        features_for_gesture_type.append(features_for_repetition)




    
    output_dir = "E:\\Chris\\putEMG Project\\extracted_features"
    os.makedirs(output_dir, exist_ok=True)  # Create the directory if it doesn't exist

    output_path = os.path.join(output_dir, 'extracted_features_subject'+ str(sub_num)+ '.mat')
    output_data = {'features': features_for_gesture_type}
    
    scipy.io.savemat(output_path, output_data)
```
